# Analysis of the Real State  Price Prediction

The goal of this project is to better understand which features are most important in predicting real state price, and how those features relate to real state price.

# 1 Price Vs Area

I used a simple linear regression model with one feature (Area) to describe Real State Price as a function of the total Area


The figure depicts the model (red) plotted against the actual data points. each blue point is represented as (X,Y) where X is the Area of the Real State and Y is its Price

This result suggests that Area has a good significant positive impact on a Real State House. However, This is clear that we need to use more than one feature(Area) to predict the Real State Price.



# 2 Kitchen Area vs Price

After using the Area as the only feature that affects the price I used the same linear regression model with another feature (Kitchen Area) to compare between them and find the most important features

The figure depicts the model (green) plotted against the actual data points. each blue point is represented as (X, Y) where X is the Kitchen Area of the Real State and Y is its Price

This result suggests that Kitchen Area has a good significant positive impact on a Real State House. However, it is less important compared to Area.



# Conclusion

I found that both of Area and Kitchen Area are important features that help in predicting the Real State Price so I will try to combine more features to improve the accuracy of the model
